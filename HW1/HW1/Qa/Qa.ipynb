```python
# code here
# first define a function for calculating histogram
# you are free to use libraries
def calc_hist(image):
    '''
    you are free to use libraries
    calculate image histogram
    input(s):
      image (ndarray): input image
    output(s):
      hist (ndarray): computed input image histogram
    '''
    #######################
    image = image.astype(np.uint8)
    hist = np.zeros(256)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            hist[image[i,j]] += 1
    
    #######################
    
    return(hist)
```

```python
# code here
# define a function (stretch) for stretching(input:image , output: stretched image)

def stretch_hist(image):
  '''
  don't use libraries
  input(s):
    image (ndarray): input image
  output(s):
    output_image (ndarray): enhanced image with histogram stretching
  '''
  output_image = image.copy()

  #######################
  f_max = np.max(output_image)
  f_min = np.min(output_image)

  output_image = np.round(((output_image - f_min) / (f_max - f_min)) * 255)
  #######################
  
  return output_image
```

```python
# code here
# define a function (Clip) for Clipping(input:image , output: Clipped image)

def clip_hist(image, min_value, max_value):
  '''
  don't use libraries
  input(s):
    image (ndarray): input image
    min_value : min value of the histogram which you wanna clip.
    max_value : max value of the histogram which you wanna clip.
  output(s):
    output_image (ndarray): enhanced image with histogram clipping
  '''
  output_image = image.copy()
  

  #######################
  output_image[output_image < min_value] = 0
  output_image[output_image > max_value] = 255
  output_image[(output_image >= min_value) & (output_image <= max_value)] = stretch_hist(output_image[(output_image >= min_value) & (output_image <= max_value)])
  #######################

  return output_image
```

## Part b

```python
# define "image1" here

#######################
image = cv2.imread("image1.jpg", 0)
#######################
```

```python
# show the image and its histogram

#######################
plt.figure()
plt.imshow(image, cmap='gray')

plt.figure()
hist = calc_hist(image)
plt.bar(np.arange(256), hist, width=1.0, edgecolor='black')
plt.xlabel('Light intensity')
plt.ylabel('Frequency')
plt.show()
#######################
```

```python
# show the streched image and its histogram

stretched_image = stretch_hist(image)
plt.figure()
plt.imshow(stretched_image, cmap='gray')

plt.figure()
hist = calc_hist(stretched_image)
plt.bar(np.arange(256), hist, width=1.0, edgecolor='black')
plt.xlabel('Light intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# show the cliped image and its histogram

cliped_image = clip_hist(image, 200, 245)
plt.figure()
plt.imshow(cliped_image, cmap='gray')

plt.figure()
hist = calc_hist(cliped_image)
plt.bar(np.arange(256), hist, width=1.0, edgecolor='black')
plt.xlabel('Light intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# use clip_hist and stretch_hist function to improve quality of the image and show it

#######################
stretched_image = stretch_hist(image)
cliped_stretched_image = clip_hist(stretched_image, 140, 240)
plt.figure()
plt.imshow(cliped_stretched_image, cmap='gray')

plt.figure()
hist = calc_hist(cliped_stretched_image)
plt.bar(np.arange(256), hist, width=1.0, edgecolor='black')
plt.xlabel('Light intensity')
plt.ylabel('Frequency')
plt.show()
#######################
```

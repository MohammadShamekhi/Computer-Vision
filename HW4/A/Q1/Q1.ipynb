### Libraries

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

### Reading the Image

```python
image = cv2.imread("q1.png", 0)
```

### Finding Threshold

```python
# You can use cv2 methods for finding this
_, thresholded = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```

### Finding Connected Components

```python
# You can use cv2 methods for finding them
num_componenst, components = cv2.connectedComponents(thresholded)
```

#### Show the result

Show each component with a random color, and write the number of components on the result image

```python
# Print number of components
print("Number of components:", num_componenst)

# Normalize labels to fit within a color range
label_img = np.uint8(255 * (components / np.max(components)))

# Convert to jet color map
colored_img = cv2.applyColorMap(label_img, cv2.COLORMAP_JET)

# Show image
plt.figure(figsize=(8, 6))
plt.imshow(colored_img)
plt.axis("off")  
plt.show()
```
